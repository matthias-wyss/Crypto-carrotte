Question 3 for staking spot carry trade strategy

For the 2021 Bull Run (2020-06-29_to_2021-11-14)

```python
from src.data.get_full_data_coinglass import load_data_for_exchange_pair,display_statistics_table, process_data_files,  compute_funding_performance_multi_exchange, CoinGlassAPI, calculate_performance_metrics,plot_carry_trade_performance,  merge_and_compute_carry_trade, calculate_period_statistics
from src.data.get_full_data_coinglass_staking import merge_and_compute_carry_trade_with_staking, calculate_period_statistics_with_staking, plot_carry_trade_performance_with_staking, display_statistics_table_with_staking
import matplotlib.pyplot as plt
import pandas as pd


#Example usage
spot_file_path = 'data/concat/spots_data_2020-06-29_to_2021-11-14.csv'
futures_file_path = 'data/concat/futures_data_2020-06-29_to_2021-11-14.csv'

separated_data = process_data_files(spot_file_path, futures_file_path)

# Now you can use the separated data for your delta neutral strategy
# For BTC:
btc_spot_df = separated_data['btc_spot']
btc_futures_df = separated_data['btc_futures']


# For ETH:
eth_spot_df = separated_data['eth_spot']
eth_futures_df = separated_data['eth_futures']


# Load staking APR data from CSV
staking_rates_df = pd.read_csv('data/lido_eth_apr_history.csv') 


# For BTC - no staking
btc_merged = merge_and_compute_carry_trade_with_staking(
    btc_spot_df, 
    btc_futures_df, 
    is_eth=False
)
btc_stats = calculate_period_statistics_with_staking(btc_merged, is_eth=False)

# For ETH - with staking using rates from dataframe
eth_merged = merge_and_compute_carry_trade_with_staking(
    eth_spot_df, 
    eth_futures_df, 
    is_eth=True, 
    staking_rates_df=staking_rates_df,
    default_staking_rate=0.036  # Fallback rate for pre-09/2022
)
eth_stats = calculate_period_statistics_with_staking(eth_merged, is_eth=True)

# Plot results
plot_carry_trade_performance_with_staking(eth_merged, eth_stats, is_eth=True, title='ETH')

# Display comparison table
display_statistics_table_with_staking(btc_stats, eth_stats)
```
